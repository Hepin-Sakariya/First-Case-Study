# Top-10 Most Voted Movies

```python
df.sort_values("numVotes",ascending=False)[["numVotes","title","year"]].head(10)
```

```python
plt.scatter(df["imdbAverageRating"],df["numVotes"],s=df["imdbAverageRating"]*20,c=df["imdbAverageRating"]*20,
            cmap="spring",alpha=0.5,label="Movies Rating")
plt.grid(linestyle=":",color="g")
plt.xlabel("IMDB Average Rating")
plt.ylabel("Number of Votes")
plt.title("Top-10 Most Voted Movies",size=15,loc="right",color="r")
plt.legend()
```

```python
ax = sns.countplot(x='genre1',data=df)
ax.bar_label(container=ax.containers[0],size=7,color="black")
plt.xticks(rotation=90)
plt.grid(linestyle=":",color="g")
plt.title("Counting of Movie Category",size=15,loc="right")
plt.show()
```

```python
plt.figure(figsize=(8,8),dpi=100)
sns.countplot(x='genre1',data=df,hue="bechdelRating")
plt.xticks(rotation=90)
plt.grid(linestyle=":",color="g")
plt.title("Counting on bechdel Rating",size=15)
plt.show()
```

# Count the movies based on bechdelRating

```python
bechdelRating = df.groupby("bechdelRating")["title"].count()
bechdelRating
```

```python
plt.pie(x=bechdelRating.values,labels=bechdelRating.index,autopct="%.2f%%",
        counterclock=False,colors=["m","r","g","y"])
sr = plt.Circle(xy=(0,0),radius=0.75,facecolor="w")
plt.gca().add_artist(sr)

plt.title("Total movies counting on bechdel rating", size=12)
plt.show()
```

```python
df.head()
```

```python
heat_corr = df.pivot_table(index="genre1",columns="bechdelRating",values="imdbAverageRating")
```

```python
plt.figure(figsize=(5,11),dpi=100)
plt.imshow(heat_corr,cmap="BrBG",alpha=1)
plt.yticks(np.arange(len(heat_corr)),heat_corr.index)
plt.xticks(np.arange(4),heat_corr.columns)
plt.colorbar()
plt.show()
```
