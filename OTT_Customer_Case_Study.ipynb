```python
df["year"]=df["year"].astype('datetime64[ns]')
```

```python
df["customer_id"]=df["customer_id"].astype(object)
```

```python
df["gender"]=df["gender"].astype('category')
```

```python
df["churn"]=df["churn"].map({0.0:"Subscribed",1.0:"Un-Subscribed"})
```

```python
df["multi_screen"]=df["multi_screen"].astype('category')
```

```python
df.info()
```

```python
df.describe()
```

```python
df1=df.drop("year",axis=1) # std is NaN 
```

```python
df1.isnull().sum()
```

# Age Distribution

```python
sns.displot(x="age",data=df1,bins=9,col="gender",hue="churn",element="step")
plt.show()
```

```python
sns.displot(x="minimum_daily_mins",y="maximum_daily_mins",data=df1,hue="gender",palette="husl")
```

```python
plt.scatter(x="minimum_daily_mins",y="no_of_days_subscribed",data=df1,c="no_of_days_subscribed"
            ,s="no_of_days_subscribed",cmap="magma",alpha=0.7)
plt.xlabel("minimum_daily_mins")
plt.ylabel("no_of_days_subscribed")

plt.grid(linestyle=":")
```

```python
sns.kdeplot(x="videos_watched",data=df1,hue="multi_screen",fill=True)
plt.grid(linestyle=":",color="green")
plt.show()
```

```python
sns.countplot(x="gender",data=df1,hue="churn",palette="Spectral")
plt.grid(linestyle=":",color="green")
```

```python
sns.stripplot(x="gender",y="no_of_days_subscribed",data=df1,hue="multi_screen",color="r",marker="o",size=3,legend="brief")
plt.grid(linestyle=":",color="orange")
```

```python
sns.jointplot(y="no_of_days_subscribed",x="age",data=df1,color="g",kind="hex")
plt.grid(linestyle=":",color="g")
```

```python
sns.pairplot(df1,vars=["weekly_mins_watched","weekly_max_night_mins"],hue="gender",palette=["orange","green"])
```
